```python
def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot 
    border is drawn
    """
    
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    # Turn off all ticks.
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    # Now re-enable visibles.
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

### Scatterplots

```python
def df_scatterplot(df, scatterx, scattery, by="region", labeler={}):
    
    """
    Creates a scatterplot, given a dataframe, names of 2 columns, a factor, 
    and a set of labels for the regions.
    """
    
    # Set up the plot area.
    figure = plt.figure(figsize=(8,8))

    # Get current axes, to plot to.
    ax = plt.gca()

    # Divide up range(0, 1) by the number of regions. 
    # (for colors).
    cs = list(np.linspace(0,1,len(df.groupby(by))))

    xlimsd={}
    ylimsd={}

    # Dicts for each region and its x, y values.
    xs={}
    ys={}

    # Dict for colours.
    cold = {}

    # For region, DataFrame.
    for k, g in df.groupby(by):

        # Pop out the last number to be a color (RGBA) for this region.
        col = cs.pop()
        c = dark2_cmap.mpl_colormap(col)
        cold[k] = c

        # Select the columns to plot.
        x = g[scatterx]
        y = g[scattery]

        # Store as {region1: [x1, x2...xn], region2...}.
        xs[k] = x
        ys[k] = y

        # Plot the region's data.
        ax.scatter(x, y, c=c, label=labeler.get(k,k), s=40, alpha=0.4);

    # Label the axes.
    ax.set_xlabel(scatterx)
    ax.set_ylabel(scattery)

    # Turn off the grid.
    ax.grid(False)

    # Add a legend.
    ax.legend()

    return ax
```

```python
df_scatterplot(df, "linoleic", "arachidic", labeler=rmap)
```

Similarly, the above plot shows that a classifier could seperate the north from Sardinia based on an olive oil's linoleic and arachidic levels.

Scatterplots, despite the usage of transparent colours, aren't great at showing the concentration of stacked points. Rugplots and marginal plots add an extra dimension to the picture by showing the distribution of data on each axis.

### Marginal Plot (hex)

Marginal plots allow you to see the distribution and areas of high intensity more easily. Libraries such as seaborn make this very easy.

```python
sns.jointplot(df["linoleic"], df["arachidic"], kind="hex");
```

However, this simple solution breaks the theme of the previous plots. It might be better to implement custom ones, especially after seeing what using rugplot out of the box does:
