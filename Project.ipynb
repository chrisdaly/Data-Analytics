Function to print out dataframes in latex table form.

```python
import StringIO #Used as buffer

def convertToLaTeX(df, alignment="c"):
    """
    Convert a pandas dataframe to a LaTeX tabular.
    Prints labels in bold, does not use math mode
    """
    numColumns = df.shape[1]
    numRows = df.shape[0]
    output = StringIO.StringIO()
    colFormat = ("%s|%s" % (alignment, alignment * numColumns))
    #Write header
    output.write("\\begin{tabular}{%s}\n" % colFormat)
    columnLabels = ["\\textbf{%s}" % label for label in df.columns]
    output.write("& %s\\\\\\hline\n" % " & ".join(columnLabels))
    #Write data lines
    for i in xrange(numRows):
        output.write("\\textbf{%s} & %s\\\\\n"
                     % (df.index[i], " & ".join([str(val) for val in df.ix[i]])))
    #Write footer
    output.write("\\end{tabular}")
    return output.getvalue()
```

## Functions

```python
def badValues(column, criteria):
    """
    Takes in a specific column name and criteria in the form of an array. The function loops
    through the criteria, checking if each value in the dataframe column is equal to that
    single value. The results are then aggregated and a logical series is returned where True
    represents rows that didn't meet a single criteria.
    """
    # Initialize boolean series to True.
    compare = 0
    
    # Calculate length of criteria array.
    l = len(criteria)
    
    # Loop through each of the criteria.
    for i in range(l):
        
        # Boolean series comparing the column values with a criteria value.
        new_compare = df[column] == criteria[i]
        
        # Add the new comparison to the old, resulting in a numerical vector.
        compare += new_compare
        
        #print "EACH = %s\n" %(each)
        #print "COMPARE =", compare, "\n"
        
    #print compare, "\n\n"
    
    # Return a logical series, where True indicates not meeting a single criteria.
    return (compare == 0)

# df[df['A12'].isin([0, 1, 2, 3])]
```

```python
def addToDict(column, rows):
    """
    Takes in a specific column name and its associated rows of bad values. The function
    extracts the names and converts them to a clean format. The names are then added to 
    the dictionary in the form of column: [names].
    """
    # Extract the names from those rows.
    names = df[rows].index
    
    # Extract just the name from pandas index object.
    names = list(names)#[0]
    
    # Add those names to the dictionary.
    bad_dict[column] = names
```

5) Create a table of the data set attributes and their associated importances in your model,
ordered from most important to least, and include this in your report.

```python
def bestParams(tree, data):
    names = list(data.columns.values)
    importance = tree.feature_importances_
    
    d = {}
    l = len(names)
    
    for i in range(l):
        name = names[i]
        d[name] = importance[i]
    
    df = DataFrame(d.items())
    df.columns = ["Attributes", "Importance"]
    df.index = df.Attributes
    
    df = df.drop(["Attributes"], 1)
    df.sort(["Importance"], ascending=False, inplace=True)

    print convertToLaTeX(np.round(df, decimals=2))
    
    return df
```
